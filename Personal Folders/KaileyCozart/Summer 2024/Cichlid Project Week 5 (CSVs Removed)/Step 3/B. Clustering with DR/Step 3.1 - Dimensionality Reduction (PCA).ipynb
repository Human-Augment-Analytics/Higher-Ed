# Dimensionality Reduction

### 1. Read in Data

```python
# Imports
import numpy as np
import pandas as pd

# Read Data
class_id_array = pd.read_csv('class_id_array.csv')
data = pd.read_csv('data.csv')
data = data.drop(columns=['track_id'])
print(class_id_array)
print(data)
```

### 2. PCA

```python
# Imports
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# PCA for Scree Plot
pca = PCA(n_components = 4)
pca.fit(data)
x_pca = pca.transform(data)

# Scree Plot
### Percentages of Variation that Each PC Accounts For
### Step Function Shows the Cumulative Contribution
plt.figure(figsize=(10, 6))
plt.bar(range(1, pca.n_components_ + 1), pca.explained_variance_ratio_, alpha=0.5, align='center')
plt.step(range(1, pca.n_components_ + 1), np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Principal Component')
plt.grid(True)
plt.show()
```

```python
# Perform PCA
n_components = 3
pca = PCA(n_components=n_components, random_state=42)
data_pca = pca.fit_transform(data)
np.savetxt('pca_transformed_data.csv', data_pca, delimiter=',')

# Save Test Data if NN
# X_test_pca = pca.transform(X_test)
# np.savetxt('X_test_pca.csv', X_test_pca, delimiter=',')
```
