```python
# Imports
import pydotplus
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import tree
import graphviz # Must install from website to use
from IPython.display import Image
import re
```

```python
# Ensure Graphviz Can Find Path
import os
os.environ["PATH"] += os.pathsep + r"C:\Program Files\Graphviz\bin"
```

```python
# Path to Data
dot_file_path = 'decision_tree_for_viz_3.dot'

# Read and Render File
with open(dot_file_path, 'r') as file:
    dot_data = file.read()

# Modify DOT to Rotate
dot_data = dot_data.replace('digraph Tree {', 'digraph Tree {rankdir=LR;')

# Customize Look
dot_data = dot_data.replace('digraph Tree {', 
                            'digraph Tree {splines=curved; edge [penwidth=2, arrowhead=none];')
dot_data = re.sub(r'graph\s*\[.*?\]', 'graph [bgcolor=white]', dot_data)

# Use Graphviz to Render with Customizations
graph = graphviz.Source(dot_data)
graph.render("decision_tree_horizontal", format='png')

# Display DT
Image(filename="decision_tree_horizontal.png")
```
